```python
# 1. Variables and Data Types
x = 10
f = 3.14
s = "hello"
b = True
print(type(x), type(f), type(s), type(b))

# 2. Control Flow
for i in range(1, 11):
    if i % 2 == 0:
        print(f"{i} is even")
    else:
        print(f"{i} is odd")

# 3. Functions
def greet(name):
    return f"Hello, {name}!"
print(greet("Alice"))

# 4. Algorithms
def linear_search(arr, target):
    for i, val in enumerate(arr):
        if val == target:
            return i
    return -1
print(linear_search([1, 2, 3, 4], 3))
def factorial_iter(n):
    result = 1
    for i in range(1, n+1):
        result *= i
    return result
print(factorial_iter(5))

# 5. Data Structures
# Stack
stack = []
stack.append(1)
stack.append(2)
print(stack.pop())
# Queue
queue = []
queue.append(1)
queue.append(2)
print(queue.pop(0))
# Dictionary
capitals = {"France": "Paris", "Italy": "Rome", "Japan": "Tokyo"}
print(capitals["Italy"])

# 6. Recursion and Iteration
def fib_recursive(n):
    if n <= 1:
        return n
    return fib_recursive(n-1) + fib_recursive(n-2)
print(fib_recursive(6))
def fib_iterative(n):
    a, b = 0, 1
    for _ in range(n):
        a, b = b, a + b
    return a
print(fib_iterative(6))
```
